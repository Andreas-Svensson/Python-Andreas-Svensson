## random module

```python
# no need to install since random already is included in standard language
import random as rnd

number_sixes = 0
number_rolls = 100000

for i in range(number_rolls):
    dice = rnd.randint(1,6) # sets dice to random integer from 1 to 6
    if (dice == 6):
        number_sixes += 1

print(f"{number_sixes = }")
print(f"Proportion of sixes: {number_sixes / number_rolls}")
```

## nested loops

```python
# range starts at first number, and ends at last number - 1, in this case 4 means it ends on 3
for row in range(1, 4): # 1, 2, 3
    for col in range(1, 4): # 1, 2, 3
        print(f"({row},{col})", end="") # end="" decides how we finish our print, by default line break, but end="" changes to print all on same line
    # in row 1, we print column 1, 2, 3, then the inner loop is finished
    print() # creates an empty new line each time all columnt have been counted through
    # then continues looping with row 2, etc..
```
