```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 20) # few points -> bad graph
f = lambda x: np.cos(x) # access element of x instead of returning all of x

f(2) # can pass in other values than x since we used lambda
```

## MATLAB style interface

```python
plt.figure(figsize = (12, 4), dpi = 45)
plt.plot(x, f(x), 'o--')
plt.title("cos(x)")
plt.xlabel("x")
plt.ylabel("y")
# plt.hlines()
```

## OOP approach

```python
x = np.linspace(-10, 10, 1000) # more points for x
```

```python
# instantiated fig, ax from figure and axes
fig, ax = plt.figure(figsize = (8, 4)), plt.axes() # fig is figure square, and in it we use ax to draw
ax.plot(x, f(x)) # ax object has a plot method
ax.set(xlabel = "x", ylabel = "y", title = "$\cos{x}$", xlim = (-5, 5), ylim = (-1.2, 1.2))
fig.savefig("assets/cos(x).png", dpi = 200) # save to file in assets folder
```

## Subplots

```python
# MATLAB style
plt.figure(figsize = (8, 4))
plt.subplot(2, 2, 1) # rows, columns, panel number
plt.plot(x, f(x))
plt.plot(x, np.sin(x)) # plotting in same subplot
plt.title("cos(x), sin(x)")
plt.xlabel("x")
plt.ylabel("y")

plt.subplot(2, 2, 2)
plt.plot(x, f(x))
plt.plot(x, -(f(x)))
plt.title("cos(x), -cos(x)")
plt.xlabel("x")
plt.ylabel("y")

plt.subplot(2, 2, 3)
plt.plot(x, -(f(x)))
plt.plot(x, -(np.sin(x)))
plt.title("-cos(x), -sin(x)")
plt.xlabel("x")
plt.ylabel("y")

plt.subplot(2, 2, 4)
plt.plot(x, np.sin(x))
plt.plot(x, -(np.sin(x)))
plt.title("sin(x), -sin(x)")
plt.xlabel("x")
plt.ylabel("y")

plt.suptitle("Trigonometric functions", fontweight = "bold");
plt.tight_layout()
```

```python
# OOP style
fig, ax = plt.subplots(1, 2, figsize = (16, 5)

ax[0].plot(x, f(x))
ax[0].plot(x, np.sin(x))

ax[1].plot(x, np.sin(x))

ax[0].set(title = "cos(x) & sin(x)")
fig.suptitle("Big title", fontsize = 25)

# if we would have 2x2 subplots, 3rd ax: ax[1][0] (row 1 col 0)
```
