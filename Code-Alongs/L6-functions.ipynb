# functions

```python
# define a function called cuber, it takes in 1 parameter
def cuber(x):
    return x**3

# calls the function cuber, using 1 argument
print(f"{cuber(5) = }")

cubes = [cuber(i) for i in range(10)]
cubes
```

```python
import matplotlib.pyplot as plt

plt.plot(cubes); # semicolon mutes last print from jupyter
```

```python
def my_mean(number1, number2, number3):
    print(f"{number1 = }, {number2 = }, {number3 = }")
    return (number1 + number2 + number3) / 3

my_mean(number2 = 1, number3 = 2, number1 = 3) # keywprd arguments
```

## default values

```python
# default parameters name and age
def hello(number_of_greetings, name = "Name", age = 0):
    for i in range(number_of_greetings):
        print(f"Hello {name}, you are {age} years old")
        # this is a void function (no returns)
        # no need to define return types in python

# positional arguments (argument 1,2,3 in order)
hello(1)
hello(1, "Andreas")
hello(3, "Andreas", 27)

# positional argument number_of_greetings
# keyword argument age
# positional arguments must go first
hello(1, age = 27)
```

```python
def word_counter(text = ""):
    word_list = text.split()
    return len(word_list)

quote = "No I am your father"
word_counter(quote)
```

## arbitrary arguments - *args

```python
def my_mean2(*numbers): # * - allows any amount of positional arguments
    sum_ = 0
    for number in numbers:
        sum_ += number
    return sum_ / len(numbers)

my_mean2(1,2,3,4,5)
```

## lambda

- anonymous function
- define a function without giving it a variable
- useful because it can be used as an argument for other functions

```python
cuber = lambda x: x**3

cuber(3)
```

## draw lines

```python
def draw_lines(x_list, k = 1, m = 1, title = ""):
    y_list = [k * x_ + m for x_ in x_list]

    plt.plot(x_list, y_list)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title(title)

x = list(range(-10,10))
draw_lines(x, k = -2, m = -1, title = "lines")

# TODO find intersection point between this and another lines
```
