# Using shapes from shapes.py

```python
# importing the specific shapes that will be used
from shapes import Circle, Rectangle, Sphere, Cuboid
import matplotlib.pyplot as plt
```

# Plotting Shapes

```python
fig, ax = plt.subplots()
c1, r1 = Circle(), Rectangle(x=2)
shape_list = [c1, r1]

for shape in shape_list:
    ax.add_patch(shape.plot())

ax.autoscale()
ax.set_aspect(1)
```

```python
fig, ax = plt.subplots()
r1 = Rectangle()

patch_list = r1.plot_translation(.5, .5)

for patch in patch_list:
    ax.add_patch(patch)

ax.autoscale()
ax.set_aspect(1)
```
