```python
# importing the specific shapes that will be used
from shapes import Circle, Rectangle, Sphere, Cuboid
import matplotlib.pyplot as plt
```

# Plotting shapes together in same plot

```python
fig, ax = plt.subplots()
c1, r1 = Circle(), Rectangle(x=2)
shape_list = [c1, r1]

for shape in shape_list:
    ax.add_patch(shape.plot())

ax.autoscale()
ax.set_aspect(1)
```

# Plotting translation of shapes together in same plot

```python
fig, ax = plt.subplots()
c1 = Circle()
r1 = Rectangle(x=2)
shape_list = [c1, r1]

for shape in shape_list:
    # create patches of each shape in their pre- and post- translation position
    patch_list = shape.plot_translation(.5, .5)
    for patch in patch_list:
        ax.add_patch(patch) # add specific patch to ax

ax.autoscale()
ax.set_aspect(1)
```
