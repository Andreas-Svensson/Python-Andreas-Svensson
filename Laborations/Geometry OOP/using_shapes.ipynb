```python
# importing the specific shapes that will be used
from shapes import Circle, Rectangle, Sphere, Cuboid
import matplotlib.pyplot as plt
import numpy as np
```

# Plotting shapes together in same plot

```python
fig, ax = plt.subplots()
c1, r1 = Circle(), Rectangle(x=2)
shape_list = [c1, r1]

for shape in shape_list:
    ax.add_patch(shape.plot())

ax.autoscale()
ax.set_aspect(1)
```

# Checking if specified shapes contains random points

```python
fig, ax = plt.subplots()
shape_list = [c1, r1, r2] = Circle(y=3), Rectangle(x=2, height=2.5), Rectangle()

# plotting all shapes:
for shape in shape_list:
    ax.add_patch(shape.plot())
ax.autoscale()
ax.set_aspect(1)
xlim, ylim = ax.get_xlim(), ax.get_ylim()

# randomizing points and checking if they are in any of the shapes:
points = 100 # amount of random points to generate
x_randoms, y_randoms = np.random.uniform(xlim[0], xlim[1], points), np.random.uniform(ylim[0], ylim[1], points)
for x, y in zip(x_randoms, y_randoms):
    check_contains = [True for shape in shape_list if shape.contains_point(x, y)]
    if True in check_contains:
        plt.plot(x, y, 'o', color = "green")
    else:
        plt.plot(x, y, 'x', color = "red")
```

# Plotting translation of shapes together in same plot

```python
fig, ax = plt.subplots()
c1 = Circle()
r1 = Rectangle(x=2)
shape_list = [c1, r1]

for shape in shape_list:
    # create patches of each shape in their pre- and post- translation position
    patch_list = shape.plot_translation(.5, .5)
    for patch in patch_list:
        ax.add_patch(patch) # add specific patch to ax

ax.autoscale()
ax.set_aspect(1)
```

```python
for shape in shape_list:
    print(f"{shape.inspect}")
```
