# Using Shapes
Example usage of shapes from shapes module with visual demonstration

```python
from shapes import Circle, Rectangle # importing shapes to use 
import matplotlib.pyplot as plt
import numpy as np
```

# Plotting shapes together in same plot

```python
fig, ax = plt.subplots()
shapes = Circle(), Rectangle(x=2)

for shape in shapes:
    ax.add_patch(shape.plot())

ax.autoscale()
ax.set_aspect(1)
```

# Checking if specified shapes contains random points

```python
fig, ax = plt.subplots()
shapes = Circle(y=3), Rectangle(x=2, height=2.5), Rectangle(), Circle(x=5, y=1.5, radius=2)

# plotting all shapes:
for shape in shapes:
    ax.add_patch(shape.plot())
ax.autoscale()
ax.set_aspect(1)

# randomizing points and checking if they are in any of the shapes:
points = 100 # amount of random points to generate
xlim, ylim = ax.get_xlim(), ax.get_ylim() # limits within which to randomize points
x_randoms, y_randoms = np.random.uniform(xlim[0], xlim[1], points), np.random.uniform(ylim[0], ylim[1], points)
for x, y in zip(x_randoms, y_randoms):
    check_contains = [True for shape in shapes if shape.contains_point(x, y)]
    if True in check_contains: # if one of shape's contains_point method returned True:
        plt.plot(x, y, 'o', color = "green")
    else: # if all of shape's contains_point methods returned False:
        plt.plot(x, y, 'x', color = "red")
```

# Plotting translation of shapes together in same plot

```python
fig, ax = plt.subplots()
shapes = Circle(), Rectangle(x=2)
move_x, move_y = .5, .5 # distance to translate shapes

for shape in shapes:
    # create patches of each shape in their pre- and post- translation position
    patches = shape.plot_translation(move_x, move_y)
    for patch in patches:
        ax.add_patch(patch) # add patches to ax

ax.autoscale()
ax.set_aspect(1)
```
