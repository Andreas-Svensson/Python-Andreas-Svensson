```python
import numpy as np
import matplotlib.pyplot as plt
```

### Creating Arrays

```python
a1 = np.array([3, 5, 7, 3])
a2 = np.zeros(3) # array of zeros of length 3
a3 = np.ones(10) # array of ones of length 10
a4 = np.random.random(10) # array of 10 random numbers that are uniformly random between 0 and 1
a5 = np.random.randn(10) # array of random numbers distributed according to gaussian distribution standard deviation of 1
a6 = np.linspace(0, 10, 100) # array of 100 equally spaced numbers from 0 to 10, NOTE: useful in plotting
a7 = np.arange(0, 10, 0.02) # array of n numbers from 0 to 10, spaced by 0.02, NOTE: useful in plotting
```

### Array Operations

element-wise operations

```python
a1
```

```python
print(a1*2) # multiplies each element of a1 by 2
print([0, 1, 2] * 2) # entire list 2 times
```
